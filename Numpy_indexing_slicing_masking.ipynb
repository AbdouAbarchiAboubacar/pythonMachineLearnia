```python
face = misc.face(gray=True)
plt.imshow(face , cmap = plt.cm.gray)
plt.show()
```

```python
axe1,axe2 = face.shape
startaxe1 = int(axe1/3)
startaxe2 = int(axe2/3)
endaxe1 = int((axe1*2)/3)
endaxe2 = int((axe2*2)/3)

newface = face[startaxe1:endaxe1,startaxe2:endaxe2]
newface[newface >200] = 255
plt.imshow(newface , cmap = plt.cm.gray)
plt.show
```
