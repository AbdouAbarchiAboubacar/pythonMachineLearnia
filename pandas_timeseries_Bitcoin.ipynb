```python
import numpy as np 
import matplotlib.pyplot as plt 
import pandas as pd 
```

```python
bitcoin = pd.read_csv('data\\BTC-EUR.csv',index_col='Date' , parse_dates=True)
bitcoin.head()
```

```python
bitcoin['Close'].plot(figsize=(12,6))
plt.show()
```

```python
# Observer l'evolution seulement en 2019
bitcoin.loc['2020','Close'].resample('M').plot()
```

### La fonction resample
#### Elle permet de regrouper les donnees selon une frequence temporelle

```python
bitcoin.loc['2020','Close'].resample('2W').mean().plot()  # W or 2W or M
```

```python
bitcoin.loc['2020','Close'].resample('M').std().plot()
```

```python
plt.figure(figsize=(12,8))
bitcoin.loc['2020','Close'].plot()
bitcoin.loc['2020','Close'].resample('M').mean().plot(label='moyenne par mois' , lw=3 , ls=':',alpha=0.8)
bitcoin.loc['2020','Close'].resample('W').mean().plot(label='moyenne par semaine' , lw=2 , ls='--',alpha=0.8)
bitcoin.loc['2020','Close'].rolling(window=7).mean().plot(label='Moving average')
bitcoin.loc['2020','Close'].rolling(window=7,center=True).mean().plot(label='Moving average with center:True')
bitcoin.loc['2020','Close'].ewm(alpha=0.6).mean().plot(label='Exp weighted')
plt.legend()
plt.show()
```

### La fonction Aggregate
#### Elle permet de rassembler dans un meme tableau plusieus statistique par desssus resample

```python
bitcoin.loc['2020','Close'].resample('W').agg(['mean','std' , 'min' ,'max'])
```

```python
m = bitcoin['Close'].resample('W').agg(['mean','std' , 'min' ,'max'])
plt.figure(figsize=(12,8))
m['mean']['2019'].plot(label='moyenne par semaine')
plt.fill_between(m.index , m['max'], m['min'] , alpha=0.2 , label='min-max par semaine')
plt.legend()
plt.show()
```

### Moving Average
#### Faire une moyenne sur une fenetre de valeur puis on decale la fenetre d'une valeur vers la droite et on calcule une nouvelle moyenne et ainsi de suite
#### on definit cette fonction <<>roulante> avec la fonction pd.rolling(window)

```python
bitcoin.loc['2020','Close'].rolling(window=7).mean().plot()
```

### Exp Weighted Function

```python
bitcoin.loc['2020','Close'].ewm(alpha=0.6).mean().plot()
```

```python
plt.figure(figsize=(12,8))
bitcoin.loc['2020','Close'].plot()
for i in np.arange(0.2,1,0.2):
    bitcoin.loc['2020','Close'].ewm(alpha=i).mean().plot(label=f'ewm {i}', ls='--' , alpha=0.8)
plt.legend()
plt.show()
```

# Assembler des datasets

```python
ethereum = pd.read_csv('data\\ETH-EUR.csv' , index_col='Date' , parse_dates=True)
```

```python
ethereum['2020']['Close'].plot()
```

```python
#Assembler
btc_eth = pd.merge(bitcoin,ethereum, on='Date' , how='inner',suffixes=('_btc','_eth')) # how='inner' :: outer : left : right
```

```python
btc_eth[['Close_btc','Close_eth']].plot(subplots=True , figsize=(12,8))
```

```python
correlations = btc_eth[['Close_btc','Close_eth']].corr()
```

```python
#Observer les correlationl
import seaborn as sns
sns.heatmap(correlations)
```
